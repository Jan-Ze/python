```python
# 导入pandas库
import pandas as pd
# 创建DataFrame对象
data = pd.DataFrame({'性别':['男', '女', '男', '女', '女'],
                     '城市':['.', '朝州市', '山头市', '汕尾市', '苏州市'],
                     '年龄':[25, 26, 24, 25, 26],
                     '爱好':['蓝球', '跑步', '跳舞', '逛街', '读书']})
# 查看data
data
```

```python
# 替换data中的错别字
dic = {'朝州市':'潮州市','山头市':'汕头市','蓝球':'篮球'}
data = data.replace(dic)
data
```

```python
import warnings

# 关闭警告显示
warnings.filterwarnings('ignore')

# 读取并查看【商品销售数据.xlsx】工作簿的数据
data = pd.read_excel('第九课/商品销售数据.xlsx')
data.head(10)
```

```python
# 将读取到的工作簿数据，进行缺失值、重复值和异常值的清洗
data.isna()
```

```python
# 清洗【用户 ID】列的缺失值
data = data.dropna(subset=['用户 ID'])
# 查看清洗后的数据基本信息总结
data.info()
```

```python
data[data.duplicated()]
```

```python
# 删除重复值
data = data.drop_duplicates()
# 查找清洗后的数据是否存在重复数据
data[data.duplicated()]
```

```python
# 查看数据描述性统计的信息
data.describe()
```

```python
# 筛选【数量】列大于 0 的数据
data = data[(data['数量'] > 0)]
# 查看数据描述性统计的信息
data.describe()
```

```python
# 计算“总金额”
data['总金额'] = data['数量'] * data['价格']
data.head(10)
```

```python
# 按【订单号】和【用户 ID】分组后，获取【发货日期】列的最大值和【总金额】列的总和
grouped_data = data.groupby(['订单号', '用户 ID'], as_index=False).agg({'发货日期': 'max', '总金额': 'sum'})
grouped_data.head(10)
```

```python
# 计算时间间隔（天数）
today = '2012-01-01 00:00:00'
pd.to_datetime(today) - pd.to_datetime(grouped_data['发货日期'])
```

```python
# 计算时间间隔
today = '2012-01-01 00:00:00'
grouped_data['时间间隔'] = (pd.to_datetime(today) - pd.to_datetime(grouped_data['发货日期'])).dt.days
grouped_data
```

```python
# 按【用户 ID】分组后，获取【时间间隔】列的最小值、【订单号】列的数量以及【总金额】列的总和
rfm_data = grouped_data.groupby(['用户 ID'], as_index=False).agg({'时间间隔':'min','总金额':'sum','订单号':'count'})
```

```python
rfm_data
```

```python
# 修改列名为：用户 ID、时间间隔、总次数和总金额
rfm_data.columns = ['用户 ID','时间间隔','总金额','总次数']
rfm_data
```

```python
import matplotlib.pyplot as plt

# 设置中文显示字体
plt.rcParams['font.family'] = ['HeiTi TC']

# 画 R 值（时间间隔）的折线图，便于后续对数据进行分组
plt.figure(figsize=(10, 8))
x = rfm_data['时间间隔'].sort_values()
y = rfm_data.index
plt.plot(x, y)
```

```python
# 定义函数按照区间划分 R 值
def caculate_r(s):
    if s <= 100:
        return 5
    elif s <= 200:
        return 4
    elif s <= 300:
        return 3
    elif s <= 400:
        return 2
    else:
        return 1
```

```python
# 对 R 值进行评分
rfm_data['R评分'] = rfm_data['时间间隔'].agg(caculate_r)
rfm_data
```

```python
# 步骤一：画 F 值（总次数）的折线图，便于后续对数据进行分组
# 步骤二：定义函数按照区间划分标记 F 值
# 步骤三：对 F 值进行评分
plt.figure(figsize=(8,8))
x = rfm_data['总次数'].sort_values()
y = rfm_data.index
plt.plot(x,y)
```

```python
# 定义函数按照区间划分 F 值
def caculate_f(s):
    if s <= 5:
        return 1
    elif s <= 10:
        return 2
    elif s <= 15:
        return 3
    elif s <= 20:
        return 4
    else:
        return 5
```

```python
# 定义函数按照区间划分 F 值
def caculate_f(s):
    if s <= 5:
        return 1
    elif s <= 10:
        return 2
    elif s <= 15:
        return 3
    elif s <= 20:
        return 4
    else:
        return 5
```

```python
# 对 F 值进行评分
rfm_data['F评分'] = rfm_data['总次数'].agg(caculate_f)
rfm_data
```

```python
# 定义函数按照区间划分 M 值
def caculate_m(s):
    if s <= 2000:
        return 1
    elif s <= 4000:
        return 2
    elif s <= 6000:
        return 3
    elif s <= 8000:
        return 4
    else:
        return 5
```

```python
# 对 M 值进行评分
rfm_data['M评分'] = rfm_data['总金额'].agg(caculate_m)
rfm_data
# 计算 R评分、F评分、M评分的平均数
r_avg = rfm_data['R评分'].mean()
f_avg = rfm_data['F评分'].mean()
m_avg = rfm_data['M评分'].mean()

print('R评分的均值为：{}，F评分的均值为{}，M评分的均值为{}'.format(r_avg, f_avg, m_avg))
```
